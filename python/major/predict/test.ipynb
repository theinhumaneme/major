```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model
import joblib
```

```python
# model = load_model("./ann-classifier.h5")
model = joblib.load("./RandomForestClassifierModelHyperTuned.sav")
```

```python
dataset = pd.read_csv("./test3.csv")

x = dataset.iloc[:,:21].values
x
```

```python
list = []
for data in x:
    d = data.tolist()
    # print(data.tolist())
    result = model.predict([d])
    # print(result.tolist())

    # FOR ANN MODEL
    # print(result.tolist()[0].index(1)+1)
    # list.append(result.tolist()[0].index(1)+1)

    # FOR ML MODEL
    print(result.tolist()[0])
    list.append(result.tolist()[0])
```

```python
import json
for i in range(0,len(list)):
    list[i] = int(list[i])
print(list)
# # with open("hello.json") as file
# json.dump(list,"hello.txt")
with open("sample.json", "w") as outfile:
    json.dump(list, outfile)
```
