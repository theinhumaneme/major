## <b>ANN Model</b>

```python
# OneHotEncoding Target Variable
y_train_encoded = pd.get_dummies(y_sam)
y_test_encoded = pd.get_dummies(y_test)
# y_train_encoded
```

```python
# Code from https://www.kaggle.com/fatihdeniz/ann-fetal-health-classification
ann_classifier = keras.Sequential([
    # Input Layer
    keras.layers.Dense(21, input_shape=X_train_scaled[0].shape, activation="relu", kernel_initializer="HeNormal"),
    
    # Hidden Layer 1
    keras.layers.Dense(40,activation="relu"),
    keras.layers.Dense(80,activation="relu"),
    keras.layers.Dropout(0.3),
    
    # Hidden Layer 2
    keras.layers.Dense(160,activation="relu"),
    keras.layers.Dense(320,activation="relu"),
    keras.layers.Dropout(0.3),
    
    # Hidden Layer 3
    keras.layers.Dense(640,activation="relu"),
    keras.layers.Dense(3, activation="softmax")
])

ann_classifier.summary()
```

```python
ann_classifier.compile(loss="categorical_crossentropy", metrics=['accuracy','Precision','Recall','AUC'], optimizer="rmsprop")

history = ann_classifier.fit(X_train_scaled, y_train_encoded, validation_data=(X_test_scaled, y_test_encoded),
                             epochs=300)
```

```python
y_pred_prob = ann_classifier.predict(X_test_scaled)
y_pred_prob
```

```python
y_pred = []

for i in y_pred_prob:
    y_pred.append(np.argmax(i) + 1)
    
y_pred = np.array(y_pred)
y_pred
```

```python
ann_accuracy = accuracy_score(y_pred, y_test)
print("ANN Accuracy :", ann_accuracy)
ann_classifier.save("./ann-classifier.h5")
```
